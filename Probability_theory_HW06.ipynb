### Задание 2
В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные:

6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1

Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение 
величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
list = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]


mean = np.mean(list)
print(f'Выборочная средняя равна {mean}')
std = np.std(list, ddof=1)
print(f'Средне кв отклонение равно {std}')
n = len(list)
alpha = 0.05
print('Используем критерий стьюдента поскольку среднекв отклонение ген совокупности неизветсно')
T1 = stats.t.ppf(1-alpha/2,df=(10-1))
X1 = mean - T1*std/n**0.5
X2 = mean + T1*std/n**0.5
print(f'Тогда истинное значение величины Х лежит в диапазоне [{X1}, {X2}]')
```

### Задание 3
Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170

Рост матерей  178, 165, 165, 173, 168, 155, 160, 164, 178, 175

Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.

```python
mum = [178, 165, 165, 173, 168, 155, 160, 164, 178, 175]
girl = [175, 167, 154, 174, 178, 148, 160, 167, 169, 170]

Xmum = np.mean(mum)
Xgirl = np.mean(girl)
print(f'Средний рост матерей и дочерей соответственно {Xmum}, {Xgirl}')
Dmum = np.var(mum, ddof=1)
Dgirl = np.var(girl, ddof=1)
print(f'Дисперсия выборки матерей и дочерей соответственно {Dmum}, {Dgirl}')
print('Воспользуемся методом интервальной оуенки для разности средних арифметических:')
print('delta +/- T(alpha/2)*S_delta')
S_delta = ((Dmum+Dgirl)/10)**0.5
print(f'S_delta = {S_delta}')
T1 = stats.t.ppf(1-0.05/2,df=2*(10-1))
print(f'T1 = {T1}')
delta = Xmum - Xgirl
print(f'delta = {delta}')
X1 = round(delta - T1*S_delta,2)
X2 = round(delta + T1*S_delta,2)
print(f'Тогда доверительный интервал разности среднего роста лежит в диапазоне [{X1}, {X2}]')
```
